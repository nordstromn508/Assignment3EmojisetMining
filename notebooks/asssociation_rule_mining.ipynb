```python
import pandas as pd
import numpy as np 
from apyori import apriori
```

```python
df = pd.read_excel('Emojisets.xlsx')
df.head()
```

```python
newdf = df['Emojiset']
newdf.head()
```

```python
X = newdf.tolist()
X
```

```python
# split X by comma
X = [x.split(',') for x in X]
X
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
```

```python
teAry = te.fit(X).transform(X)
```

```python
ddf = pd.DataFrame(teAry, columns=te.columns_)
ddf.head()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

test = apriori(ddf, min_support=0.02, use_colnames=True)
```

```python
test
```

```python
# Collecting the inferred rules in a dataframe
rules = association_rules(test, metric ="lift", min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
print(rules.head(10))
```

```python
# get rules into table
rules.to_excel('rules.xlsx')
```
